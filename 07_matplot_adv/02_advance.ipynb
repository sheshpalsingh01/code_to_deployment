```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ensure x is defined
x = np.linspace(0, 1, 100)

# use plt.subplots (plural) to create multiple axes
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Line Plot")

axes[1].plot(x, x**2, 'r--', x, np.exp(x), 'g^',x, np.log(x+1), 'bs',x, np.sin(x), 'k:')
axes[1].set_title("Line Plot with different style")
plt.show()
```

```python
x=np.linspace(0,1,100)
fig,axe=plt.subplots()
axe.plot(x,np.cos(2*np.pi*x),'r--',label='cosine',lw=2)
axe.set_xticks([0,0.25,0.5,0.75,1])
axe.set_xticklabels(['0','/2','','3/2','2'])
axe.set_yticks([-1,-0.5,0,0.5,1])
axe.set_yticklabels(['-1','-0.5','0','0.5','1'])
axe.set_title('Cosine Wave')
axe.set_xlabel('x axis')
plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, np.sin(5 * np.pi * x), 'b-',x, np.cos(5 * np.pi * x), 'r--')
ax.set_title('Sine and Cosine Waves')
ax.set_xticklabels(['0', '/2', '', '3/2', '2'])
ax.set_yticklabels(['-1', '-0.5', '0', '0.5', '1'])
ax.set_xlabel('x axis')

from matplotlib import ticker
formatter=ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
plt.legend(['Sine','Cosine'])
plt.grid()
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['xtick.major.pad']=5
matplotlib.rcParams['xtick.major.pad']=5
x=np.linspace(0,1,100)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x,np.sin(5*np.pi*x),'b-',x,np.exp(x),'r--')
ax.set_title('Sine and Cosine Waves')
ax.set_xticklabels(['0','/2','','3/2','2'])
ax.set_yticklabels(['-1','-0.5','0','0.5','1'])
# padding between axis label and axis numbers
ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 5
ax.set_xlabel('x axis')
plt.grid()
plt.tight_layout()
plt.show()
```
