### Fancy Indexing

```python
arr_2d=np.zeros((10,10))
arr_2d
```

```python
arr_lenght=arr_2d.shape[1]
arr_lenght
```

```python
for i in range(arr_lenght):
    arr_2d[i]=i
    
arr_2d
```

```python
arr_2d[[2,3,4,5]]
```

```python
## allows in any order
arr_2d[[2,4,6,1]]
```
