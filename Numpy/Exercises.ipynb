```python
n = 390 
drift = 0.0002
vol = 0.01
returns = np.random.normal(drift, vol, n)
price = 100 * np.exp(np.cumsum(returns))
```

```python
price
```

```python
## Vectorizedd bid

bids=100+np.random.normal(0,0.05,n)
asks=bids+np.random.uniform(0.01,0.1,n)
```

```python
## Slippage and order Impact Simulation
valume=np.random.randint(100,1000,n)
impact=np.random.uniform(0,0.002,n)
execution_price=price*(1+impact*(valume/valume.max()))
execution_price
```

```python
## Moving Avg Crossover
## Detect buy/sell signals efficiently.
short_ma=np.convolve(price, np.ones(10)/10, mode='valid')
long_ma=np.convolve(price, np.ones(50)/50, mode='valid')
signals=np.where(short_ma[40:] > long_ma, 1, 0)
signals
```

```python
## Z-Score Mean Reversion
## Identify extreme price deviations
window = 20
rolling_mean = np.convolve(price, np.ones(window)/window, 'valid')
rolling_std = np.sqrt(np.convolve((price[window-1:] - rolling_mean)**2, np.ones(window)/window, 'valid'))
z_score = (price[window*2-2:] - rolling_mean) / rolling_std
z_score
```
