### Multi-index and Index Hierarchy

```python
# Index Levels and Hierarchy
outside=['G1','G1','G1','G2','G2','G2']
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
```

```python
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
```

```python
df
```

```python
df.loc['G1']
```

```python
df.loc["G1"].loc[1]
```

```python
df.loc['G1']["A"]  # access row C
```

```python
df.index.names
```

```python
df.index.names=['Group','Num']
```

```python
df
```

```python
df.xs('G1')  # cross section
```

```python
df1=df.xs('G1')
df1["A"]
```

```python
df1.loc[1]
```
